```python

import numpy as np
import math
import sys


t = (3, 3)
a = [(2,3),(2,4), (2,2),(3,2), (4,3),(3,4)]
a = np.asarray(a)


class Point():
    def __init__(self, x, y ):
        self.x = x
        self.y = y
        self.pos = (x,y)

    def distance2(self, p):
        return (self.x - p.x)*(self.x - p.x) + (self.y - p.y)*(self.y - p.y)
    
    def distance(self, p):
        return math.sqrt(self.distance2(p))

    def __repr__(self): 
        return f"{type(self).__name__}: {self.pos}"

    @classmethod
    def from_list(cls, lst):
        return cls(lst[0], lst[1])


class Unit(Point):

    def print_pos(self):
        print(f"{type(self).__name__} : {self.pos}", file=sys.stderr, flush=True)
    
    def print_move(pos1, pos2): 
        dir = degToCompass(pos1, pos2)
        print(dir)


class Hero(Unit):
    def __init__(self, x, y):
        super().__init__(x,y)


class Monster(Unit):

    def centroid(self):
        length = self.pos.shape[0]
        sum_x = np.sum(self.pos[:, 0])
        sum_y = np.sum(self.pos[:, 1])
        return sum_x/length, sum_y/length




def angle( pos1, pos2):
    dif_y, dif_x = pos2[1] - pos1[1], pos2[0] - pos1[0]
    angl =  math.degrees(math.atan2(dif_y, dif_x))
    if angl < 0 :
        return 360 + angl
    return angl


def degToCompass(pos1, pos2):
    num = angle(pos1, pos2)
    val=int(num/45)
    arr=["E","NE","N", "NW", "W","SW","S","SE"]
    return arr[(val % 8)]
```
